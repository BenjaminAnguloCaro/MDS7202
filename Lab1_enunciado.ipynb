Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?

2. Nombre y explique 3 diferencias entre `Git` y `Github`.

3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.

4. ¿Que elementos componen a la estructura de datos de `Git`?

5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

<code>Escriba sus respuestas acá:</code> 

1- Git es un sistema de control de versiones ya que permite a sus usuarios mantener un registro de los cambios realizados en un repositorio además de permitir la colaboración entre varios desarrolladores, permitiendo así tener un control sobre las versiones de un proyecto, por ejemplo, al mantener la rama "main" para producción y las demás ramas para desarrollo de nuevas funciones, solución de problemas, etc. 

2- En primer lugar, se tiene que Git es un sistema de control de versiones, mientras Github es un servicio de alojamiento basado en web. Es decir, Git permite gestionar y controlar cambios en el código fuente de un proyecto, mientras Github permite a los desarrolladores almacenar, colaborar y compartir sus proyectos utilizando Git.\
En segundo lugar, Git es una herramienta de línea de comandos, y por su parte, GitHub proporciona una interfaz gráfica. Es decir, al interactuar con Git se utilizan en la terminal comandos como 'git commit', 'git push', 'git pull', entre otros; en cambio, la interacción con Github permite a los desarrolladores realizar diversas operaciones de Git de manera visual.\
Por último, Git crea un repositorio local para trackear los cambios locales, y Github es un espacio para subir copias del repositorio. De esta forma, Git funciona localmente en la máquina de un desarrollador; GitHub, por otro lado, es una plataforma en línea que proporciona un servidor remoto para almacenar repositorios Git.

3- Un "hash" es un valor generado a través de una función de hashing, la cual convierte un valor en un valor alfanumérico que no se puede revertir a su forma original. Ahora en lo referente al commit, el hash es un representativo de este mismo, tomando en consideración los archivos commiteados, el mensaje del commit, entre otros valores. Un ejemplo del hash de un commit es este: 3f7e6bfc6387935f73260b46e260a9aa07ce754e el cual representa al commit realizado al responder la pregunta 1 de esta parte del laboratorio. 

4- La estructura de datos de Git está compuesta por blobs, trees, commits y hashs. Los blobs son objetos donde se almacena el contenido de un archivo, los árboles corresponden a directorios para blobs y más árboles, y los commits representan un conjunto de cambios en el repositorio en un momento específico. Por su parte, un Hash puede apuntar a un blob, un commit o un árbol. Este cuenta con 40 caracteres, pero sólo unos pocos suelen ser suficientes para identificar un commit.
5- Para crear un repositorio es necesario hacer uso del comando "git init" para crear un repositorio en la carpeta actual sobre la que se esté ubicado. Luego de esto el repositorio se puede ir llenando con los comandos "git add" y "git commit" para añadir los cambios y guardarlos. Por último, para hacer un seguimiento de los cambios en el repositorio se puede hacer uso de "git status" para ver la cantidad de commits realizados en el branch sobre el que estemos o también se puede usar "git diff" para ver los cambios específicos en los distintos archivos del repositorio hasta el último commit. 

Responda si las siguientes afirmaciones son Verdaderas (V) o Falsas (F). Justifique en **cualquiera** de los casos (0.2 pts c/u):


6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.
7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.
8. `git status` genera los mismos resultados que `git log`.
9. En `Git` es posible deshacer cambios y volver a versiones anteriores del código.
10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.

<code>Escriba sus respuestas acá:</code>

6- Falso. Git no siempre resuelve automáticamente los conflictos de líneas al momento de hacer merge entre diferentes ramas. A veces, cuando hay cambios conflictivos en las mismas líneas de código, se generan problemas que deben ser resueltos manualmente.

7- Falso, el comando 'git pull' se usa para poder actualizar los archivos que se tengan de manera local con respecto a los archivos que se encuentren en el repositorio remoto en la nube. En cambio el comando 'git clone' sirve para crear una copia de un repositorio nuevo en un computador, más no para actualizar uno ya existente. 

8- Falso. 'git status' muestra el estado actual del repositorio (archivos modificados, añadidos, etc.), mientras que 'git log' muestra el historial de commits del repositorio.

9- Verdadero. Esto se puede lograr usando, por ejemplo, los comandos 'git reset' (para deshacer cambios de commits anteriores) o 'git revert' (para revertir un commit dejando constancia de esto al crear un nuevo commit).
10- Falso. 'git commit' se utiliza para confirmar los cambios en el repositorio local. Para enviar los cambios al repositorio remoto, se utiliza 'git push'

Por último, investigue y comente para qué sirven los siguientes comandos:

11. `git stash` (0.25 puntos) 

    R: Sirve para poder guardar temporalmente los cambios de un archivo sin tener que realizar un commit. Esto puede ser útil por ejemplo si se desea realizar un cambio de branch sin hacer un commit de los cambios realizados en la rama actual, por lo que usando este comando se podría realizar este cambio de branch sin ningún problema para luego en un futuro poder retomar los cambios realizados. 

12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)
'git reset' y 'git revert' son dos comandos en Git que se utilizan para deshacer cambios en un repositorio, pero tienen diferencias en cuanto a cómo hacen estos cambios y cómo afectan la historia del repositorio. En el caso de 'git reset', este se utiliza para deshacer cambios locales en el repositorio y modifica la historia eliminando commits. Por su parte, 'git revert' se utiliza para deshacer cambios al crear un nuevo commit que revierte los cambios realizados por un commit anterior, de manera que mantiene la historia intacta.
